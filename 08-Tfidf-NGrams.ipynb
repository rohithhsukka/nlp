```python
import pandas as pd 
messages=pd.read_csv('SpamClassifier-master/smsspamcollection/SMSSpamCollection', sep='\t',names=["label","message"])
```

```python
messages
```

```python
## Data cleaning and preprocessing 

import re 
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordlemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA_Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review = [wordlemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Create TF-IDF and NGrams

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
```
